```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
```

```python
# Load the training data
train_df = pd.read_csv('../data/train.csv')

# Quick check
print(f"Dataset shape: {train_df.shape}")
```

```python
# Display first 5 rows
train_df.head()
```

```python
# Statistical summary of numerical columns
train_df.describe()
```

```python
# Count missing values in each column
missing_values = train_df.isnull().sum()
missing_percentage = round((train_df.isnull().sum()/len(train_df))*100, 2)

missing_df = pd.DataFrame({
    'missing_count': missing_values,
    'percentage': missing_percentage
})

print("Missing Values Analysis:")
print(missing_df[missing_df['missing_count'] > 0].sort_values('missing_count', ascending=False))
```

```python
# Check data types of each column
print("Data Types:")
print(train_df.dtypes)
```

```python
# Overall Survival Rate
print("Overall Survival Rate:")
print(f"Survived: {train_df['Survived'].sum()} ({train_df['Survived'].mean()*100:.2f}%)")
print(f"Died: {(train_df['Survived']==0).sum()} ({(1-train_df['Survived'].mean())*100:.2f}%)")
```

```python
# Passenger Class Distribution
print("Passenger Ticket Class Distribution:")
print(train_df['Pclass'].value_counts().sort_index())

print("\nSurvival by Ticket Class:")
print(train_df.groupby('Pclass')['Survived'].agg(['count', 'sum', 'mean']))
```

```python
# Gender Analysis
print("Gender Distribution:")
print(train_df['Sex'].value_counts())

print("\nSurvival by Gender:")
print(train_df.groupby('Sex')['Survived'].agg(['count', 'sum', 'mean']))
```

```python
# Embarked Port Analysis
print("Embarkation Port Distribution:")
print(train_df['Embarked'].value_counts())

print("\nSurvival by Embarkation Port:")
print(train_df.groupby('Embarked')['Survived'].agg(['count', 'sum', 'mean']))
```

```python
# Age Groups
print("Age Statistics:")
print(f"Youngest passenger: {train_df['Age'].min():.2f} years")
print(f"Oldest passenger: {train_df['Age'].max():.2f} years")
print(f"Average age: {train_df['Age'].mean():.2f} years")
print(f"Median age: {train_df['Age'].median():.2f} years")
```
